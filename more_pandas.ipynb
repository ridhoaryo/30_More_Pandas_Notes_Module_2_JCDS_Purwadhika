```python
df_dummy_1 = pd.DataFrame([
    {'nama':'Andi', 'fisika':90, 'biologi':78},
    {'nama':'Budi', 'fisika':95, 'biologi':80},
    {'nama':'Caca', 'fisika':90, 'biologi':82}])

df_dummy_2 = pd.DataFrame([
    {'nama':'Caca', 'math':90, 'kimia':78},
    {'nama':'Andi', 'math':95, 'kimia':80},
    {'nama':'Budi', 'math':90, 'kimia':82}])
df = pd.concat([df_dummy_1, df_dummy_2], axis=1, join='inner', ignore_index=True, sort=False)
df
```

```python
# agar join berfungsi, maka dataframe harus memiliki index
df_dummy_1 = pd.DataFrame([
    {'nama':'Andi', 'fisika':90, 'biologi':78},
    {'nama':'Budi', 'fisika':95, 'biologi':80},
    {'nama':'Caca', 'fisika':90, 'biologi':82}], index=[0,1,2])

df_dummy_2 = pd.DataFrame([
    {'nama':'Caca', 'math':90, 'kimia':78},
    {'nama':'Andi', 'math':95, 'kimia':80},
    {'nama':'Budi', 'math':90, 'kimia':82}], index=[2,0,1])
df = pd.concat([df_dummy_1, df_dummy_2], axis=1, join='inner', ignore_index=True, sort=False)
df
```

## 3. Merge

```python
df_dummy_1 = pd.DataFrame([
    {'nama':'Andi', 'fisika':90, 'biologi':78},
    {'nama':'Budi', 'fisika':95, 'biologi':80},
    {'nama':'Caca', 'fisika':90, 'biologi':82}])

df_dummy_2 = pd.DataFrame([
    {'nama':'Caca', 'math':90, 'kimia':78},
    {'nama':'Andi', 'math':95, 'kimia':80},
    {'nama':'Budi', 'math':90, 'kimia':82}])
df = pd.merge(df_dummy_1, df_dummy_2)
df
```

```python
df_dummy_1 = pd.DataFrame([
    {'nama':'Andi', 'fisika':90, 'biologi':78},
    {'nama':'Budi', 'fisika':95, 'biologi':80},
    {'nama':'Caca', 'fisika':90, 'biologi':82}])

df_dummy_2 = pd.DataFrame([
    {'nama':'Caca', 'math':90, 'kimia':78},
    {'nama':'Andi', 'math':95, 'kimia':80},
    {'nama':'Budi', 'math':90, 'kimia':82},
    {'nama':'Deni', 'math':90, 'kimia':82}])
df = pd.merge(df_dummy_1, df_dummy_2, on='nama', how='outer')
df
```

## 4. Pivot 

```python
df_dummy = pd.DataFrame([
    {'test':1, 'nama':'Andi', 'fisika':90, 'biologi':78},
    {'test':2, 'nama':'Andi', 'fisika':95, 'biologi':80},
    {'test':3, 'nama':'Andi', 'fisika':90, 'biologi':82},
    {'test':1, 'nama':'Budi', 'fisika':80, 'biologi':80},
    {'test':2, 'nama':'Budi', 'fisika':85, 'biologi':89},
    {'test':3, 'nama':'Budi', 'fisika':72, 'biologi':75},])
df_dummy
```

```python
df_dummy_1 = df_dummy.pivot(index='test', columns='nama')
df_dummy_1
```
